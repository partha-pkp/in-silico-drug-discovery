```python
import numpy as np  # Import NumPy for numerical operations
import pandas as pd  # Import Pandas for data manipulation and analysis
from sklearn.ensemble import RandomForestClassifier  # Import Random Forest classifier
from sklearn.preprocessing import StandardScaler  # Import scaler for feature normalization

target = "senolytic"  # Define the target column name (label we want to predict)

# Separate features (X) and target (y)
X = data_set_2.drop(columns=[target]).copy()  # Remove target column from dataset to create feature matrix
y = data_set_2[target].copy()  # Extract target column as labels

# Convert categorical variables into numerical format using one-hot encoding
X = pd.get_dummies(X, drop_first=True)  # drop_first=True avoids multicollinearity

# Replace infinite values with NaN (to clean the dataset)
X = X.replace([np.inf, -np.inf], np.nan)  # Handle infinities in features
y = y.replace([np.inf, -np.inf], np.nan)  # Handle infinities in target

# Ensure all data is numeric (convert invalid entries to NaN)
X = X.apply(pd.to_numeric, errors="coerce")  # Convert all feature columns to numeric
y = pd.to_numeric(y, errors="coerce")  # Convert target to numeric

# Fill missing values in features with column mean
X = X.fillna(X.mean())  # Impute missing values using mean of each column

# Remove rows where target is missing
valid_rows = y.notna()  # Identify rows where target is not NaN
X = X.loc[valid_rows]  # Keep only valid rows in features
y = y.loc[valid_rows]  # Keep only valid rows in target

# Convert target to integer (required for classification)
y = y.astype(int)  # Ensure target labels are integers (classes)

# Clip extreme values to float32 range to avoid overflow issues
max_float32 = np.finfo(np.float32).max  # Get maximum float32 value
X = X.clip(lower=-max_float32, upper=max_float32)  # Limit values within float32 range

# Normalize features (zero mean, unit variance)
scaler = StandardScaler()  # Initialize scaler
X_scaled = scaler.fit_transform(X)  # Fit scaler and transform data

# Convert scaled array back to DataFrame (preserve column names and indices)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)

# Initialize Random Forest classifier
model = RandomForestClassifier(
    n_estimators=500,  # Number of trees in the forest
    random_state=42,   # Fix randomness for reproducibility
    n_jobs=-1          # Use all CPU cores for faster training
)

# Train the model
model.fit(X_scaled, y)  # Fit model using scaled features and labels

# Extract feature importance scores
rf_importance = pd.DataFrame({
    "Feature": X.columns,  # Feature names
    "Score": model.feature_importances_  # Importance scores from Random Forest
}).sort_values(by="Score", ascending=False).reset_index(drop=True)  # Sort descending

# Print feature ranking
print("\nFeature Ranking (Random Forest Classification)")
for i, row in rf_importance.iterrows():  # Loop through ranked features
    print(f"{i+1}. {row['Feature']} : {row['Score']:.6f}")  # Print rank, feature name, and score
```

```python
import matplotlib.pyplot as plt

# Set Georgia font globally
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Georgia"]

# Select top features
top_n = 20
top_features = rf_importance.head(top_n)

# Plot (only blue)
plt.figure(figsize=(12, 6))

plt.bar(top_features["Feature"], top_features["Score"], color="blue")

plt.xticks(rotation=45, ha="right")
plt.xlabel("Feature")
plt.ylabel("Feature Importance (Random Forest)")
plt.title(f"Top {top_n} Feature Importance (Random Forest)")

plt.tight_layout()

# Save as JPG
plt.savefig("rf_feature_importance.jpg", dpi=300, bbox_inches="tight")

plt.show()
```

```python
import numpy as np  # Import NumPy for numerical operations
import pandas as pd  # Import Pandas for data manipulation
import shap  # Import SHAP for model explainability
from sklearn.ensemble import RandomForestClassifier  # Random Forest classifier
from sklearn.preprocessing import StandardScaler  # Feature scaling

target = "senolytic"  # Define the target variable (label)

X = data_set_2.drop(columns=[target]).copy()  # Remove target column to get features
y = data_set_2[target].copy()  # Extract target column

X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables into numeric (one-hot encoding)

X = X.replace([np.inf, -np.inf], np.nan)  # Replace infinite values in features with NaN
y = y.replace([np.inf, -np.inf], np.nan)  # Replace infinite values in target with NaN

X = X.apply(pd.to_numeric, errors="coerce")  # Convert all feature values to numeric (invalid → NaN)
y = pd.to_numeric(y, errors="coerce")  # Convert target to numeric

X = X.fillna(X.mean())  # Replace NaN in features with column mean

valid_rows = y.notna()  # Identify rows where target is not NaN
X = X.loc[valid_rows]  # Keep valid rows in features
y = y.loc[valid_rows]  # Keep valid rows in target

y = y.astype(int)  # Convert target to integer labels (required for classification)

max_float32 = np.finfo(np.float32).max  # Maximum float32 value
X = X.clip(lower=-max_float32, upper=max_float32)  # Limit values to avoid overflow

scaler = StandardScaler()  # Initialize scaler
X_scaled = scaler.fit_transform(X)  # Fit and transform features
X_scaled = pd.DataFrame(X_scaled, columns=X.columns, index=X.index)  # Convert back to DataFrame

model = RandomForestClassifier(
    n_estimators=200,  # Number of trees in the forest
    random_state=42,   # For reproducibility
    n_jobs=-1          # Use all CPU cores
)

model.fit(X_scaled, y)  # Train the model

explainer = shap.TreeExplainer(model)  # Create SHAP explainer for tree-based model
shap_values = explainer.shap_values(X_scaled)  # Compute SHAP values

if isinstance(shap_values, list):
    # Case 1: Multi-class classification → list of arrays (one per class)
    mean_abs_shap = np.mean([np.abs(sv).mean(axis=0) for sv in shap_values], axis=0)

else:
    # Case 2: Single output or unified SHAP format
    if len(shap_values.shape) == 3:
        # Shape: (samples, features, classes)
        mean_abs_shap = np.abs(shap_values).mean(axis=(0, 2))
    else:
        # Shape: (samples, features)
        mean_abs_shap = np.abs(shap_values).mean(axis=0)

shap_importance = pd.DataFrame({
    "Feature": X.columns,  # Feature names
    "Score": mean_abs_shap  # Mean absolute SHAP importance
}).sort_values(by="Score", ascending=False).reset_index(drop=True)  # Sort descending

print("\nFeature Ranking (SHAP) - Random Forest Classification")
for i, row in shap_importance.iterrows():  # Iterate through ranked features
    print(f"{i+1}. {row['Feature']} : {row['Score']:.6f}")  # Print rank, feature, score
```

```python
import matplotlib.pyplot as plt

# Set Georgia font globally
plt.rcParams["font.family"] = "serif"
plt.rcParams["font.serif"] = ["Georgia"]

# Select top features
top_n = 20
top_features = shap_importance.head(top_n)

# Plot (only green)
plt.figure(figsize=(12, 6))

plt.bar(top_features["Feature"], top_features["Score"], color="green")

plt.xticks(rotation=45, ha="right")
plt.xlabel("Feature")
plt.ylabel("Mean Absolute SHAP Value")
plt.title(f"Top {top_n} SHAP Feature Importance")

plt.tight_layout()

# Save as JPG
plt.savefig("shap_feature_importance_green.jpg", dpi=300, bbox_inches="tight")

plt.show()
```
